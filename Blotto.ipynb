```python
import numpy as np
import itertools
from matplotlib import pyplot as plt
rng = np.random.default_rng()
epochs = 50000

k = 15  # total units
N = 5   # number of bins

dictionary = []
# weak compositions of k into N parts
for dividers in itertools.combinations(range(k + N - 1), N - 1):
    # compute sizes from adjacent divider positions
    prev = -1
    vec = []
    for d in dividers:
        vec.append(d - prev - 1)
        prev = d
    vec.append(k + N - 1 - prev - 1)
    dictionary.append(vec)
n = len(dictionary)
dictionary = np.array(dictionary)
batch = 100
#graph  = [[] for _ in range(n)]
strategy = np.ones(n) / n
cumulative_regret = np.zeros(n)

def normalformgame(idt, idc):
    t = dictionary[idt]
    c = dictionary[idc]
    return np.sum((t > c) )#* np.arange(1, N + 1))

def regretcalculation(idc):
    counterfactual_scores = np.sum((dictionary > idc)) #* np.arange(1, N + 1), axis=1)
    regrets = np.maximum(0, counterfactual_scores)
    return regrets
def sample_action(strategy):
    r = np.random.random()
    cum = np.cumsum(strategy)
    return np.searchsorted(cum, r)
```

```python
rounds = 50000
total = 0

input1 = 1
input2 = 2
input3 = 3
input4 = 4
input5 = 5

t = np.array([input1, input2, input3, input4, input5])

for i in range(rounds):
  action = rng.choice(dictionary, size=1, p=strategy)
  score = np.sum((t > action[0]))
  opponent = np.sum((t < action[0]))

  total += (score - opponent)

print("Net AVG: " + str(total / rounds))

```
